# the libraries


```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import Input , Dense , concatenate , Dropout
from tensorflow.keras.models import Model
```

```python
# image data
image_data = np.random.rand(100,30)

# sensor data

sensor_data = np.random.rand(100,5)

# binary labels

labels = np.random.randint(0,2,100)
```

```python
image_input = Input(shape=(30,),name="image_input")
sensor_input = Input(shape=(5,),name="sensor_input")

x1 = Dense(20,activation = 'relu')(image_input)
x1 = Dropout(0.2)(x1)
x1 = Dense(10, activation = 'relu')(x1)


x2 = Dense(20 , activation = 'relu')(sensor_input)
x2 = Dropout(0.2)(x2)
x2 = Dense(10 , activation = 'relu')(x2)

merged = concatenate([x1,x2])

z = Dense(10, activation = 'relu')(merged)
z = Dropout(0.2)(z)
output = Dense(1 , activation = 'sigmoid')(z)

model = Model(inputs=[image_input,sensor_input],outputs = output)
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history = model.fit([image_data,sensor_data],labels,epochs=10,batch_size=18,validation_split=0.2,verbose=0)
```

```python
loss, accuracy = model.evaluate([image_data,sensor_data],labels)

print(f"loss:{loss}, accuracy:{accuracy}")


sample_image = np.random.rand(50,30)
sample_sensor = np.random.rand(50,5)

prediction = model.predict([sample_image,sample_sensor])

print(f"prediction: {prediction}")
```
